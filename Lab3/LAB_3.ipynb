```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Hàm ánh xạ sử dụng sin thay vì cos
def GeoFun(outcoord):
    a = 10 * np.sin(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.sin(outcoord[1] / 10.0) + outcoord[1]
    return a, b

# Đọc ảnh mức xám
data = iio.imread('world_cup.jpg', mode='F')

# Áp dụng biến đổi với size = (5, 5)
d1 = nd.geometric_transform(data, GeoFun, output_shape=(5, 5))
#output_shape=(5, 5) tạo ảnh rất nhỏ (5x5 pixel), nên khi hiển thị sẽ mờ, mất chi tiết. 
# Khi ảnh chỉ có 5x5 pixel, dữ liệu quá ít → gần như toàn giá trị 0 (đen hoàn toàn).
# Hiển thị ảnh
plt.imshow(d1, cmap='gray')
plt.title("Biến dạng sin với size = (5, 5)")
plt.axis('off')
plt.show()
```
