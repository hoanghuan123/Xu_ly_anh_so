```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Hm nh x s dng sin thay v cos
def GeoFun(outcoord):
    a = 10 * np.sin(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.sin(outcoord[1] / 10.0) + outcoord[1]
    return a, b

# c nh mc xm
data = iio.imread('world_cup.jpg', mode='F')

# p dng bin i vi size = (5, 5)
d1 = nd.geometric_transform(data, GeoFun, output_shape=(5, 5))
#output_shape=(5, 5) to nh rt nh (5x5 pixel), nn khi hin th s m, mt chi tit. 
# Khi nh ch c 5x5 pixel, d liu qu t  gn nh ton gi tr 0 (en hon ton).
# Hin th nh
plt.imshow(d1, cmap='gray')
plt.title("Bin dng sin vi size = (5, 5)")
plt.axis('off')
plt.show()
```
